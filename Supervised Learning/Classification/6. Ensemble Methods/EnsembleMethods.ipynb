# Ensemble Methods

Here we will look at two classification ensemble methods: Bagging and random forest.

## Algorithms
### Bagging

Bagging, also known as Bootstrap aggregating, takes in a training dataset and splits it into $n$ datasets (one for each estimator) sampled uniformly and with replacement from the original dataset. Then, for each estimator, a simple classification algorithm is trained on the subset. Then, predictions can be made by giving each estimator a "vote" and picking the classification most commonly voted upon.

### Random Forest

The random forest algorithm is a subset of the Bagging algorithm, when used with decision tree model estimators. However, each estimator only looks at a random subset of the features of the training data. Otherwise, the rest of the algorithm is identical to the Bagging algorithm.

### Bagging Advantages

- More accurate than the individual component models
- Simple due to simple component models

### Bagging Disadvantages

- Component models are correlated
- More expensive than the single models

### Random Forest Advantages

- More accurate than the individual decision trees
- Simple due to decision tree simplicity
- Prevents creating artificial correlation between the individual decision trees

### Random Forest Disadvantages

- More expensive than the individual decision trees
- Sometimes overfits

```python
data_dir = "/home/patrick/Projects/DataScienceAndMachineLearning/Data/Titanic-Dataset"
train_X = np.load(f"{data_dir}/train_X.npy")
train_y = np.load(f"{data_dir}/train_y.npy")
test_X = np.load(f"{data_dir}/test_X.npy")
test_y = np.load(f"{data_dir}/test_y.npy")
PCA_train_X = np.load(f"{data_dir}/PCA_train_X.npy")
PCA_train_y = np.load(f"{data_dir}/PCA_train_y.npy")
PCA_test_X = np.load(f"{data_dir}/PCA_test_X.npy")
PCA_test_y = np.load(f"{data_dir}/PCA_test_y.npy")
train_X
```

```python
classifiers = []
predictions = []
n_vals = [1, 10, 50, 100, 150, 200, 500]
for n in n_vals:
    bagging_decision_tree_2_classifier = BaggingClassifier(DecisionTreeClassifier(max_depth = 2, random_state=42), n_estimators = n, random_state=42)
    bagging_decision_tree_5_classifier = BaggingClassifier(DecisionTreeClassifier(max_depth = 5, random_state=42), n_estimators = n, random_state=42)
    bagging_decision_tree_15_classifier = BaggingClassifier(DecisionTreeClassifier(max_depth = 15, random_state=42), n_estimators = n, random_state=42)
    bagging_logistic_regression_classifier = BaggingClassifier(LogisticRegression(), n_estimators=n, random_state=42)
    bagging_knn_classifier = BaggingClassifier(KNeighborsClassifier(), n_estimators=n)
    random_forest_classifier = RandomForestClassifier(n_estimators=n, random_state=42)
    decision_tree_2_classifier = DecisionTreeClassifier(max_depth=2, random_state=42)
    decision_tree_5_classifier = DecisionTreeClassifier(max_depth=5, random_state=42)
    decision_tree_15_classifier = DecisionTreeClassifier(max_depth=15, random_state=42)

    bagging_decision_tree_2_classifier.fit(train_X, train_y)
    bagging_decision_tree_5_classifier.fit(train_X, train_y)
    bagging_decision_tree_15_classifier.fit(train_X, train_y)
    bagging_logistic_regression_classifier.fit(train_X, train_y)
    bagging_knn_classifier.fit(train_X, train_y)
    random_forest_classifier.fit(train_X, train_y)
    decision_tree_2_classifier.fit(train_X, train_y)
    decision_tree_5_classifier.fit(train_X, train_y)
    decision_tree_15_classifier.fit(train_X, train_y)

    bagging_decision_tree_2_prediction = bagging_decision_tree_2_classifier.predict(test_X)
    bagging_decision_tree_5_prediction = bagging_decision_tree_5_classifier.predict(test_X)
    bagging_decision_tree_15_prediction = bagging_decision_tree_15_classifier.predict(test_X)
    bagging_logistic_regression_prediction = bagging_logistic_regression_classifier.predict(test_X)
    bagging_knn_prediction = bagging_knn_classifier.predict(test_X)
    random_forest_prediction = random_forest_classifier.predict(test_X)
    decision_tree_2_prediction = decision_tree_2_classifier.predict(test_X)
    decision_tree_5_prediction = decision_tree_5_classifier.predict(test_X)
    decision_tree_15_prediction = decision_tree_15_classifier.predict(test_X)

    classifiers.append([bagging_decision_tree_2_classifier, bagging_decision_tree_5_classifier, bagging_decision_tree_15_classifier, bagging_logistic_regression_classifier,
                        bagging_knn_classifier, random_forest_classifier, decision_tree_2_classifier, decision_tree_5_classifier, decision_tree_15_classifier])
    
    predictions.append([bagging_decision_tree_2_prediction, bagging_decision_tree_5_prediction, bagging_decision_tree_15_prediction, bagging_logistic_regression_prediction,
                        bagging_knn_prediction, random_forest_prediction, decision_tree_2_prediction, decision_tree_5_prediction, decision_tree_15_prediction])
```

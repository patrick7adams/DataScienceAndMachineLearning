# Data Preprocessing
In this file, we will preprocess and prepare the 'Auto-MPG' dataset for use with the rest of the classification data.

Then, we will split this data into training and testing data.

### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn

import struct

sns.set_theme()
```

### Data Explanation
- mpg (target) - the miles traveled per gallon of gas for the given car, a measure of travel efficiency
- displacement (feature) - the total volume of air and fuel moved by the engine's pistons per cycle for the given car, measured in cubic centimeters
- cylinders (feature) - The number of cylinders that the given car's engine has
- horsepower (feature) - The horsepower of the engine of the given car
- weight (feature) - The weight of the given car.
- acceleration (feature) - The maximum acceleration of the given car.
- model_year (feature) - The initial production year of the given car model.
- is_USA (feature) - Whether the given car was produced in the USA. 1 if it was, 0 if not.
- is_Europe (feature) - Whether the given car was produced in the Europe. 1 if it was, 0 if not.
- is_Japan (feature) - Whether the given car was produced in the Japan. 1 if it was, 0 if not.

### Data Preprocessing
First, we will load in the data, assigning columns accordingly:

```python
data_dir = "/home/patrick/Projects/DataScienceAndMachineLearning/Data/"

names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
df = pd.read_table(data_dir + 'auto-mpg.data', sep = r'\s+', names=names)
df
```

Drop the car name column:

```python
df = df.drop(columns=["car_name"])
df
```

Now, one hot encode the origin column, as it is categorical with only three entries:

```python
df["is_USA"] = (df["origin"] == 1).astype(np.int64)
df["is_Europe"] = (df["origin"] == 2).astype(np.int64)
df["is_Japan"] = (df["origin"] == 3).astype(np.int64)
df
```

Replace '?' entries in the horsepower column with the mean horsepower, so that we will not have to drop those columns:

```python
df["horsepower"] = df["horsepower"].replace("?", np.nan).astype(np.float64)
meanval = df["horsepower"].mean()
df["horsepower"] = df["horsepower"].fillna(meanval)
df
```

Finally, drop remaining useless columns:

```python
df = df.drop(columns=["origin", "model_year"])
df = df.astype(np.float64)
df
```

```python
for col in df.columns:
    df[col] = (df[col] - df[col].mean()) / df[col].std()
df
```

### Train / Test Split
First, set train/test split parameters:

```python
train_test_ratio = 0.7
train_cutoff = round(train_test_ratio * df.shape[0])
```

Now, perform the split:

```python
X_data = df.iloc[:, 1:].to_numpy()
y_data = df.iloc[:, 0].to_numpy()
train_X = df.iloc[:train_cutoff, 1:].to_numpy()
test_X = df.iloc[train_cutoff:, 1:].to_numpy()
train_y = df.iloc[:train_cutoff, 0].to_numpy()
test_y = df.iloc[train_cutoff:, 0].to_numpy()
train_X
```

Output data:

```python
data_dir = "/home/patrick/Projects/DataScienceAndMachineLearning/Data/AutoMPG-Dataset"
np.save(f"{data_dir}/train_X.npy", train_X)
np.save(f"{data_dir}/train_y.npy", train_y)
np.save(f"{data_dir}/test_X.npy", test_X)
np.save(f"{data_dir}/test_y.npy", test_y)
```
